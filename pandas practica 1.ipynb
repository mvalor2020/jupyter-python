```python
#crea una nueva columna concatenar dos culunas que existen en la base de datos(--solo variables str)
bd1['nueva columna'] = bd1['NOMBRE_PACIENTE'] +' '+ bd1['PATOLOGIA AGRUPADA']
```

```python
bd1
```

```python
#borramos columna pasa
bd1= bd1.drop(['nueva columna'], axis=1)
```

```python
bd1.dtypes
```

```python
#eliminar caracteres "letras texto" del documento que debe ser numeros

def dejar_solo_cifras(txt):
  return "".join(c for c in txt if c.isdigit())
bd1.DOCUMENTO = bd1.DOCUMENTO.map(dejar_solo_cifras)
```

```python
## cambia el tipo de object a int
bd1['DOCUMENTO'] = pd.to_numeric(bd1['DOCUMENTO'])
```

```python
## cambia el tipo de object a cadena
bd1['ESTANCIA'] = pd.to_numeric(bd1['DOCUMENTO'])
```

```python
## cambia el tipo de object a category
cats = [category for category in bd1.ESTANCIA.unique()]
bd1['ESTANCIA'] = bd1.ESTANCIA.astype(pd.api.types.CategoricalDtype(categories = cats, ordered=True))
```

```python
bd1.dtypes
```

```python
# sacar año y mes en columnas apartes de una fecha(datatime)
bd1['FECHA_INGRESO'] = pd.to_datetime(bd1['FECHA_INGRESO'])
bd1['year'], bd1['month'] = bd1['FECHA_INGRESO'].dt.year, bd1['FECHA_INGRESO'].dt.month
bd1
```

```python
bd1.dtypes
```

```python
bd1['month'].unique()
```

```python
## cambiar el tipo de variable int object
bd1[['month']] = bd1[['month']].astype(object)
```

```python
bd1.dtypes
```

```python
# recategorizar datos de una columna
bd1.loc[bd1['month'] == 10, 'month'] = 'Octubre'
bd1.loc[bd1['month'] == 1, 'month'] = 'Enero'
bd1.loc[bd1['month'] == 2, 'month'] = 'Febrero'
bd1.loc[bd1['month'] == 3, 'month'] = 'Marzo'
bd1.loc[bd1['month'] == 5, 'month'] = 'Mayo'
bd1.loc[bd1['month'] == 6, 'month'] = 'Junio'
bd1.loc[bd1['month'] == 7, 'month'] = 'Julio'
bd1.loc[bd1['month'] == 8, 'month'] = 'Agosto'
bd1.loc[bd1['month'] == 9, 'month'] = 'Septiembre'
bd1.loc[bd1['month'] == 11, 'month'] = 'Noviembre'
bd1.loc[bd1['month'] == 12, 'month'] = 'Diciembre'
bd1.loc[bd1['month'] == 4, 'month'] = 'abril'
bd1
```
